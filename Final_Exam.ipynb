```python
alldb_file = os.path.join(path,'alldb.csv')
```

```python
data_alldb = pd.read_csv(alldb_file)
data_alldb
```

```python
data_alldb.isnull().any() # check ค่าว่าง
```

```python
data_sub = data_alldb[['user', 'RowKey', 'date', 'phone', 'site', 'time', 'urine', 'ov', 'PartitionKey', 'title', 'age']] #เลือกคอลัมน์ที่ต้องการ
data_sub
```

#user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)


```python
alldb_user= data_sub[(data_sub['user'].astype(str).str.isnumeric()) & (data_sub['user'].str.len() == 5)] # สร้างตัวแปร alldb_user มารับโดยเลือกที่ตาราง data_sub ที่คอลัมน์ user โดยเปลี่ยนค่า
                                                                                                                                                # ข้อมูลในนั้นให้เป็น string และใช้คำสั่ง .isnumeric() เชคว่าเป็นตัวเลขหรือไม่ และรวมกับ
                                                                                                                                                # ข้อมูลคอลัมน์ user ที่เป็นตัวเลข ความยาว 5 ตัว
alldb_user
```

```python
alldb_user.isnull().any()
```

# site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb_site= alldb_user[(alldb_user['site'].astype(str).str.isnumeric()) & (alldb_user['site'].str.len() == 5)] # สร้างตัวแปร alldb_site มารับโดยเลือกที่ตาราง alldb_user ที่คอลัมน์ site โดยเปลี่ยนค่า
                                                                                                                                                # ข้อมูลในนั้นให้เป็น string และใช้คำสั่ง .isnumeric() เชคว่าเป็นตัวเลขหรือไม่ และรวมกับ
                                                                                                                                                # ข้อมูลคอลัมน์ site ที่เป็นตัวเลข ความยาว 5 ตัว
alldb_site
```

```python
alldb_site.isnull().any()
```
