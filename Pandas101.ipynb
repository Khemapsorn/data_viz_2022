1คำนวณ อายุเฉลี่ย **ผู้ชาย** ในข้อมูลทั้งหมด

````
this_data = data_covid[['sex','age','province_of_onset']] #กำหนดตัวแปร this_data และเลือกคอมลัมน์ ที่ต้องการใช้ได้แก่ sex age province_of_onset 
````

````
this_data #แสดงข้อมูลที่เลือก
````

````
male = this_data[this_data['sex']=='ชาย'] #กำหนดตาราง male ให้ใช้ข้อมูลในตาราง this_data ที่มีเงื่อนไขว่าคมลัมน์ sex ต้องเป็นชาย 
male['age'] # เลือกคอลัมน์ age ของเเพศชายในตาราง male
````

````
sum = 0 #กำหนด sum เป็น 0 เผื่อใช้รวมผลลัพธ์ใน loop |ตัวแปร sum จะมารับค่าทั้งหมด  ถ้าเริ่มที่1 คือเอาตัวแปรมาคูณกัน เอกลักษณ์ การบวก การคูณ
N = 0 #กำหนด N เป็น 0 เพื่อใช้นับจำนวนผลลัพธ์ใน loop 
for m in male['age']: # ให้ m วนloop ในตาราง male ที่ คอมลัน์ age
    
    if m > 0: # ถ้ามันมากกว่า 0 แสดงว่าเป็นอายุคนปกติ | ถ้า m ในตาราง male ที่คอลัมนื age >0 ให้:
        sum += m #ให้ m บวกเพิ่มเข้าไปใน sum | sum = sum + m  ผลรวมอายุทั้งหมด
        N += 1 # ให้ N บวกเพิ่มขึ้นทีละ 1 |N บวกเพิ่มทีละ1  จำนวนคนทั้งหมด

print(f'อายุเฉลี่ย ของ ผู้ชาย {sum/N}') #แสดงผลลัพธ์ที่เป็นค่าเฉลี่ยของ sum/N
````

2.คำนวณ อายุเฉลี่ย  **ผู้ชาย** ในจังหวัดขอนแก่น

````
male_KK = male[male['province_of_onset']=='ขอนแก่น'] #กำหนดตัวแปร male_KK ให้เลือกตาราง male ที่คอลัมน์ provint_of_onset เป็นขอนแก่นเท่านั้น

sum = 0 #กำหนด sum เป็น 0 เผื่อใช้รวมผลลัพธ์ใน loop
N = 0 #กำหนด N เป็น 0 เพื่อใช้นับจำนวนผลลัพธ์ใน loop
for m in male_KK['age']: # ให้ m วนloop ในตาราง male_KK ที่ คอมลัน์ age
    
    if m > 0: #ถ้าตาราง male ที่คอลัมนื age >0 ให้:
        sum += m # sum = sum + m |ให้ m บวกเพิ่มเข้าไปใน sum
        N += 1 # ให้ N บวกเพิ่มขึ้นทีละ 1

print(f'อายุเฉลี่ย ของ ผู้ชาย ในจังหวัดขอนแก่น {sum/N}') #แสดงผลลัพธ์ที่เป็นค่าเฉลี่ยของ sum/N
````
